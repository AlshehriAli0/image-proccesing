# Lab 5 – Spatial Filtering: Smoothing and Sharpening
**ARTI 404 – Image Processing**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import data
```

## Task 1: Unsharp Masking

```python
image = data.chelsea()
sigma = 2
amount = 1.5

image_float = image.astype(np.float32) / 255.0
blurred = cv2.GaussianBlur(image_float, (0, 0), sigmaX=sigma, sigmaY=sigma)
sharpened = np.clip(image_float + amount * (image_float - blurred), 0, 1)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")
ax[1].imshow(blurred)
ax[1].set_title("Blurred Image")
ax[1].axis("off")
ax[2].imshow(sharpened)
ax[2].set_title("Sharpened (Unsharp Masking)")
ax[2].axis("off")
plt.show()
```

## Assessment

### Assessment Task 1: 7x7 Box Filter

```python
image = data.camera()
kernel = np.ones((7, 7), np.float32) / 49
smoothed = cv2.filter2D(image, -1, kernel)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original")
ax[0].axis("off")
ax[1].imshow(smoothed, cmap='gray')
ax[1].set_title("7x7 Box Filter")
ax[1].axis("off")
plt.tight_layout()
plt.show()
```

### Assessment Task 2: Gaussian Filter (5x5 and 21x21)

```python
image = data.astronaut()
gaussian_5 = cv2.GaussianBlur(image, (5, 5), 0)
gaussian_21 = cv2.GaussianBlur(image, (21, 21), 0)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(image)
ax[0].set_title("Original")
ax[0].axis("off")
ax[1].imshow(gaussian_5)
ax[1].set_title("5x5 Gaussian")
ax[1].axis("off")
ax[2].imshow(gaussian_21)
ax[2].set_title("21x21 Gaussian")
ax[2].axis("off")
plt.tight_layout()
plt.show()
```

### Assessment Task 3: Laplacian Sharpening

```python
image = cv2.cvtColor(data.coffee(), cv2.COLOR_RGB2GRAY)
image_float = image.astype(np.float32) / 255.0

laplacian = cv2.Laplacian(image_float, cv2.CV_32F, ksize=3)
sharpened = np.clip(image_float - laplacian, 0, 1)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(image_float, cmap='gray')
ax[0].set_title("Original")
ax[0].axis("off")
ax[1].imshow(laplacian, cmap='gray')
ax[1].set_title("Laplacian")
ax[1].axis("off")
ax[2].imshow(sharpened, cmap='gray')
ax[2].set_title("Sharpened")
ax[2].axis("off")
plt.tight_layout()
plt.show()
```
